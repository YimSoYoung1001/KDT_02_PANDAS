## (1) 결측값 해결하기

```python
#age
print(titanic['age'].describe(), titanic['age'].median(), sep='\n\n')               
# ==> 결측값을 28로 채우기로 결정
```

```python
titanic['age'] = titanic['age'].astype('str')
titanic['age'] = titanic['age'].str.strip()
titanic['age'].unique()
```

```python
titanic['age'] = titanic['age'].replace('nan', '28')
titanic['age'].unique()       #결측값 사라짐!
```

```python
titanic['age'] = titanic['age'].astype('float64')
titanic.info()
```

```python
#embarked
titanic['embarked'].describe(include="all")         
# ==> 결측값을 'S'로 채움
```

```python
titanic['embarked'] = titanic['embarked'].astype('str')
titanic['embarked'] = titanic['embarked'].str.strip()
titanic['embarked'].unique()
```

```python
titanic['embarked'] = titanic['embarked'].replace('nan', 'C')
titanic['embarked'].unique()   
```

```python
titanic['age'] = titanic['age'].astype('object')
titanic.info()
```

```python
#deck
titanic['deck'].describe(include="all")      
```

```python
titanic['deck'] = titanic['deck'].astype('str')
titanic['deck'] = titanic['deck'].str.strip()
titanic['deck'].unique()
```

```python
titanic['deck'] = titanic['deck'].replace('nan', 'C')
titanic['deck'].unique()   
```

```python
titanic['deck'] = titanic['deck'].astype('object')
titanic.info()
```

```python
#embark_town
titanic['embark_town'].describe(include="all")   
```

```python
titanic['embark_town'] = titanic['embark_town'].astype('str')
titanic['embark_town'] = titanic['embark_town'].str.strip()
titanic['embark_town'].unique()
```

```python
titanic['embark_town'] = titanic['embark_town'].replace('nan', 'Southampton')
titanic['embark_town'].unique()   
```

```python
titanic['embark_town'] = titanic['embark_town'].astype('object')
titanic.info()
```
