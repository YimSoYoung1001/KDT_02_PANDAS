```python
#1 
netflix.info()
```

```python
netflix = pd.read_csv(file, parse_dates = ['date_added'])
netflix['type'] = netflix['type'].astype('category')
netflix.info()
```

```python
#2 제목이 Limitless인 모든 콘텐츠 찾기
netflix[netflix["title"] == 'Limitless']
```

```python
#3 감독 Robert Rodriguez 유형 Movie인  모든 콘텐츠 찾기
dir = netflix["director"] == 'Robert Rodriguez'
cat = netflix["type"] == 'Movie'
netflix[dir&cat]
```

```python
#4 2019-07-31 이거나 Robert Altman
date = netflix['date_added'] == '2019-07-31'
dir = netflix['director'] == 'Robert Altman'
netflix[date|dir]
```

```python
#5 감독이 'Orson Welles'  또는   'Aditya Kripalani'  또는   'Sam Raimi' 
dir1 = netflix['director'] == 'Orson Welles'
dir2 = netflix['director'] == 'Aditya Kripalani'
dir3 = netflix['director'] == 'Sam Raimi'
netflix[dir1 | dir2 | dir3]
```

```python
#6 2019-05-01    ~    2019-06-01 인 모든 콘텐츠
netflix[netflix['date_added'].between('2019-05-01', '2019-06-01')]
```

```python
#7 director에서 NaN 값이 있는 모든 행 삭제
netflix.dropna(subset = 'director')
```

```python
#8 콘텐츠를 단 하나만 등록한 날짜를 식별
# netflix.drop_duplicates(subset = ['date_added'])             #이것은 중복되면 1개는 살아남아있다.
netflix.drop_duplicates(subset = ['date_added'], keep = False)
```
