[문1] 아래 조건을 만족하는 Series를 생성하세요.
1) 데이터 : 1~50 범위에서 7의 배수 정수
2) 인덱스 : 'a~z' 범위에서 데이터 갯수 만큼의 알파벳
3) 타  입 : 부호가 없는 1바이트 정수
4) 이  름 : sevens
5) 속  성 : index, values, dtype, ndim, shape, size

```python
#1 데이터를 리스트로 만들었음
list7 = []
for i in range(1,51):
    if i % 7 == 0:
        list7.append(i)
list7
```

```python
import pandas as pd
import numpy
```

```python
#2 #3 #4  시리즈를 만들었음 (인덱스 설정, 타입 설정, 이름 설정)
indexnum = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
sevens = pd.Series(data = list7, index = indexnum, name = 'sevens', dtype = 'int8')
sevens
```

```python
sevens.info()
```

```python
# 5 속성 추출함
print(sevens.index, sevens.values, sevens.dtype, sevens.ndim, sevens.shape, sevens.size, sep='\n')
```

[문4] 아래 조건을 만족하는 데이터 객체를 생성하세요.
1) 데이터 : stock price.xlsx
2) 데이터 로딩 시 인덱스로 사용할 수 있는 컬럼 설정
3) 결측치 처리 =>
   (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
   (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
4) value가 1000이하 이거나 price가 90000이상인 데이터

```python
#1 #2 데이터를 읽고 인덱스로 사용할 컬럼을 지정함
file = '../DATA/stock price.xlsx'
file = pd.read_excel(file, index_col='id')
file
```

```python
#3 결측치 처리 - (가) 1개라도 NaN이면 모두 제거 / price가 NaN이면 제거 / value가 10000이하면 제거
file1 = file.dropna()
x = file1['value'] > 10000
file1[x]
```

```python
#3 결측치 처리 - (나) 평균값으로 결측치 대체, 이전/이후 값으로 결측치 대체
file.describe()
```

```python
file.info()
```

```python
#대체할 결측치 변수 (평균값) 설정
valueN = '81263.944755'
valueP = '127667.5000'

# 결측치를 평균값으로 변환
import numpy as np
file['value'] = file['value'].replace(np.nan, valueN)
file['price'] = file['price'].replace(np.nan, valueP)

# 결측값이 바뀌었는가 확인
file
```

```python
file['value'] = file['value'].astype('float64')
file['price'] = file['price'].astype('float64')
```

```python
#4 value가 1000이하 이거나 price가 90000이상인 데이터 추출
data1 = file['value'] <= 1000
data2 = file['price'] >= 90000
file[data1 | data2]
```
