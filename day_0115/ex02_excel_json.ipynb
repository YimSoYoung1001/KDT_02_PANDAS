## [JSON ==> DataFrame 저장하기]

(1) 데이터 준비

```python
file3 = '../DATA/nobel.json'
```

(2) JSON ==> DataFrame으로 저장

```python
nobelDF = pd.read_json(file3)
nobelDF.info()
```

(3) 데이터 확인

```python
print(nobelDF.head(3), nobelDF.tail(3), sep='\n\n')
```

(4) 데이터 추출

```python
# 행 추출 => iloc[] / loc[]
nobelDF.iloc[0]                #행 1줄을 뺐다. 뺀거는 시리즈의 형태
```

```python
nobelDF.iloc[0]['prizes']      #행 1줄을 뺀 상태에서 레이블을 불러서 데이터를 불렀다. 그것은 딕셔너리의 형태
```

```python
# 행 추출 => iloc[[n1, ... ,]]/loc[]
nobelDF.iloc[[0,3]]       
```

```python
oneData =  nobelDF.iloc[0]['prizes']
type(oneData)
```

```python
#oneData가 딕셔너리 형태니까 아래와 같이 접근 가능하다.
print(oneData.keys(), oneData.items(), sep='\n\n')
```

```python
for k, v in oneData.items():
    print(f'[{k}] \n===> {v}')
    print()
    if k == 'laureates':
        for d in v:            # 여기서 v는 dict 원소로 구성된 list
            for k1, v1 in d.items():
                print("-", k1, v1)
```

- json  중첩 데이터의 다중 반복문 ==> 단일 리스트 형태 처리 : 평탄화 / 정규화

```python
pd.json_normalize(oneData)
# 여기서 'laureates'에서 각 아이디를 컬럼명으로 하고 그 밸류를 데이터로 하기위해 아래와 같이 함
```

```python
pd.json_normalize(oneData, record_path='laureates')
#어떤 완벽한 데이터를 레코드로 하여 데이터 프레임으로 뽑을거냐 => record_path
```

```python
pd.json_normalize(oneData, record_path='laureates', meta=["year", "category"])
# meta 변수지정을 통해 시라졌던 year, category가 컬럼명에 추가되었다.

#이렇듯 늘 데이터 원본을 먼저 보아야 무엇을 제어하고 다루어야할지 보인다. 무턱대로 부르면 ㄴㄴ
```

- 모든 행에 대한 정규화/평탄화 진행

```python
# 컬럼 읽기 => DF객체변수명['컬럼명'], DF객체변수명.컬럼명
# ===> 모든 행에 record_path에 해당하는 키(key)가 존재해야함 
# ===> 없는 행에 record_path에 해당하는 키(key) 추가

#pd.json_normalize(nobelDF.prizes, record_path='laureates')
#KeyError 발생: "Key 'laureates' not found.

for data in nobelDF.prizes:
    print(type(data), data.keys(), end='\n\n')
    # 해당 dict에 laureates 키가 없다면 추가
    data.setdefault('laureates', [])                  #기본값을 설정하여 없다면 저렇게 빈리스트 설정해줌 
```

```python
nobelDF2 = pd.json_normalize(nobelDF.prizes, record_path='laureates', meta = ['year', 'category'])
# 위에서 없는거는 만들어놓았기때문에 더이상 keyError ㄴㄴ

nobelDF2.info()
```

(7) JSON의 다양한 방식의 DataFrame 저장하기 => orient 파라미터

```python
json_data='''
[
{"id" : "1","name":"sravan","age":22},
{"id" : "2","name":"harsha","age":22},
{"id" : "3","name":"deepika","age":21},
{"id" : "4","name":"jyothika","age":23}
]
'''
```

- orient = 'record' 설정 => JSON 파일 데이터를 행 단위로 읽어서 DataFrame에 저장

```python
df = pd.read_json(json_data, orient = 'records')
# 딕셔너리 속 키가 컬럼명이 되고 밸류가 데이터가 된다.
print(df)
```

```python
# 데이터 => 키 : {값}
json_data2 = '''
{
    "student-1" : {"id" : "1","name":"sravan","age":22},
    "student-2" : {"id" : "2","name":"harsha","age":22},
    "student-3" : {"id" : "3","name":"deepika","age":21},
    "student-4" : {"id" : "4","name":"jyothika","age":23}

}
'''
```

```python
df2 = pd.read_json(json_data2, orient = 'index')
# 키를 인덱스명으로 지정함 (orient의 default는 column)
# 키가 인덱스 레이블이 되고, 밸류속 키와 밸류는 컬럼명과 데이터가 된다.
print(df2)
```

```python
df2 = pd.read_json(json_data2).T
# 디폴트로는 오리엔트가 컬럼인데 여기에 T 메소드를 주면 인덱스와 컬럼이 서로 바뀐다.
print(df2) 
```

```python
# JSON Format 
json_data3 = '''
[
    [ "1", "sravan",22 ],
    [ "2", "harsha",22 ],
    [ "3", "deepika",21 ],
    [ "4", "jyothika",23 ]
]
'''

# 키가 없고 값만 존재하고있음
```

```python
df = pd.read_json(json_data3)     #orient = 'values'
print(df)
# 1개의 레코드가 1개의 행으로 들어감
```
