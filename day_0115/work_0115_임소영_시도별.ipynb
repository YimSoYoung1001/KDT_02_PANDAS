```python
import pandas as pd
```

```python
file = '../DATA/시도별 전출입 인구수.xlsx'
city = pd.read_excel(file)
city
```

```python
city.info()
```

```python
# -의 데이터는 정수 0으로 바꿈 (원본 변경)
city = city.replace('-', '0')
```

```python
# 형변환하기 전 - (1) 0번 row의 글자들을 없앤다.
list = []
for i in range(50):
    list.append(0)
city.iloc[0] = list
```

```python
# 형변환하기 전 - (2) Nan을 처리해준다.
city = city.fillna(0)
```

```python
# 형변환
for num in range(1970, 2018):
    num = str(num)
    city[num] = city[num].astype('int64')

city
```

## 전국에서 각 지역별로 전입한 것 분석

```python
# 원본 데이터 프레임에서 전국 부분만 꺼낸다.
city_all = city.iloc[1:19]
city_all
```

```python
# 수치 데이터의 통계값 확인
city_all.describe()
```

```python
# 최빈값 확인
city_all.mode(numeric_only = True)
```

```python
# 중앙값 확인
city_all.median(numeric_only = True)
```

```python
# 각 연도 사이의 관계를 비교함
city_all.corr(numeric_only = True)
```
