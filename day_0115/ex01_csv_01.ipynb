## [ csv 데이터 파일 다루기 ]

(1) 모듈 로딩 <hr>

```python
import pandas as pd
```

(2) 데이터 준비 <hr>

- 파일 이름 => 경로 + 파일명
- 경로 => 2가지 종류
    * 절대경로 : 드라이브부터 파일까지의 전체 경로 (ex: C:\Users\kdp\Desktop\강의 자료)
    * 상대경로 : 현재 파일을 기준으로 경로 설정  
          - ./ : 현재 위치 의미  
                  (예: ./aFile.csv  ,  aFile.csv => 현재 파일과 동일한 폴더/위치에 존재)  
          - ../ : 한 단계 위로 의미    
                  (예: ../Data/bFile.csv => 현재 파일이 존재하는 폴더 밖에 존재하는 파일)

```python
file = './test.csv'          # file = 'text.txt'와 동일 의미         (둘다 현재 위치에 있는 파일을 불러들인다)
```

```python
file2 = '../DATA/movies.csv'
```

(5) read_csv 파일 매개변수 살펴보기 <hr>

- header 파라미터 : DataFrame의 컬럼명 설정하는 파라미터
    * 기본값 : csv 파일의 첫번째 라인의 데이터를 컬럼명으로 설정
    * None : csv 파일의 첫번째 라인이 컬럼명 아님을 의미 => 컬럼명 미지정
    * 행번호 : csv 파일의 행번호 라인이 컬럼명으로 설정, 그 이전 행 번호 데이터는 사라짐!

```python
pd.read_csv(file2, header = None)  
#header 의 기본 값은 첫줄이 컬럼명이 되는데 이건 None으로 설정해서 인덱스로 컬럼명이 됨
#첫줄은 데이터
```

```python
#pd.read_csv(file2, header = 5)        #5번 줄을 컬럼명으로 설정함. 그 앞 데이터는 손실됨
```

- index_col 파라미터 : 컬럼 중에서 행 인덱스로 설정, set_index()와 동일 역할 수행

```python
pd.read_csv(file2)
```

```python
pd.read_csv(file2, index_col = 'Rank')                 #단일 인덱스        
#pd.read_csv(file2, index_col = ['Rank', 'Title'])      #멀티 인덱스                 

#하나의 가로데이터(레코드,엔티티..)는 부를때 오로지 하나만 나오게끔, 그래서 보통 숫자인거를 많이함 그러면 하나만 서치되니까
#Rank를 인덱스로 설정함
#멀티인덱스 하고 싶다면 리스트 안에 한다.
```

- sep / delimiter : csv 파일에서 데이터 구분자 설정
    * 기본값 : 쉼표(,)

```python
pd.read_csv(file2, sep=';')
#원본데이터는 쉼표로 구분되어있으니까 세미콜론으로 구분할게없다. 그래서 통으로 결과가 나옴
```

```python
#pd.read_csv(file)          #csv라 기본이 쉼표이기 때문에 그냥 구분할게 없다.
#pd.read_csv(file, header = None, sep=';')    # 방금csv 파일에 제목을 추가했다. 그래서 그상태로 다시 실행시켠 에러가 뜬다
```

- skiprows: csv 파일에 읽어 들이지 않을 행 개수를 지정하기 => 앞부분
- skipfooter: csv 파일에 읽어 들이지 않을 행 개수를 지정하기 => 끝부분

```python
pd.read_csv(file, sep=';', skiprows = 3, header = None)    
#세미콜론으로 구분하겠다
#초반 3줄을 스킵하겠다
#컬럼명은 따로 지정하지 않겠다
```

```python
pd.read_csv(file, sep=';', skiprows = 3, header = None, skipfooter = 2, engine='python') 

#끝 2줄은 읽기 않겠다
#엔진은 ParserWarning을 해결하기 위해 덧붙였다. 
```

- usecols : 컬럼에서 특정 컬럼만 선택해서 로딩
    * usecols = [0, 2, ...]
    * usecols = ['컬럼명1', '컬렴명n']

```python
pd.read_csv(file, sep=';', skiprows=3, header = None, usecols = [0,2])
```

```python
#이 둘은 같음
#pd.read_csv(file2, usecols = [0,4])
pd.read_csv(file2, usecols = ['Rank', 'Year'])
```
