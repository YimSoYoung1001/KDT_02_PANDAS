```python
# 인덱스 위치로 행 추출 => iloc[정수입력]
nba.iloc[300]
```

```python
# iloc을 이용하여 여러 행을 추출   => 리스트 활용하기
nba.iloc[[100, 200, 300, 400]]
```

```python
# iloc에 슬라이싱 활용하기 
# 슬라이싱과 유사하게 end에서 마지막 갑은 포함 X !!
# 음수도 사용가능

# 인덱스번호 400, 401, 402, 403 추출하기
nba.iloc[400:404]         #end값은 미포함
```

```python
# nba의 처음 10개 행 중에 간격이 2씩 떨어져 있는 모든 행 가져오기
nba.iloc[0:10:2]
```

```python
# 특정 열에서 값 추출 => loc 사용
# loc, iloc는 모두 두번쨰 인수로 추출할 열을 받음
# loc 사용시 열이름을 입력,  iloc 사용시 열위치를 입력

# Team 열과 Giannis Antetokoumpo 행의 교차점에 있는 값 가져오기
nba.loc["Giannis Antetokounmpo", "Team"]
```

```python
# 여러개의 값 추출
# loc 접근자에 하나 혹은 2개의 리스트를 전달

# 'James Harden' 인덱스 레이블이 있는 행과 Positon, Birthday 열에 있는 값을 추출하기
nba.loc[ "James Harden", ["Position", "Birthday"] ]
# 시리즈로 반환됨
```

```python
# 행 여러개, 열 여러개
nba.loc[ ["Russell Westbrook", "Anthony Davis"], ["Team", "Salary"] ]
```

```python
# 행 여러개, 열 여러개  +  리스트 활용하기
# 보여지는 순서에 맞춰 열의 이름을 전달해야함

result1 = nba.loc["Joel Embiid", "Position":"Salary"]
result2 = nba.loc["Joel Embiid", "Salary":"Position"]       #행 순서를 식별하지 못하기 떄문에 빈 결과가 반환

print("result1 => ", result1, sep='\n', end= '\n\n')
print("result2 => ", result2, sep='\n')
```

```python
# 특정 열에서 값 추출 => iloc 사용
# loc, iloc는 모두 두번쨰 인수로 추출할 열을 받음

# 인덱스 57에 있는 행과 인덱스 3("Salary")에 있는 열의 교차점에 있는 값 추출하기
nba.iloc[57, 3]
```

```python
# iloc에 슬라이싱 활용하기
nba.iloc[100:104, :3]
```

```python
# iloc, loc은 유연성이라는 장점과 오버헤드라는 단점이 있음 (판다스는 사용자가 어떤 종류의 입력을 했는지 알아야함)
# at, iat을 통해서도 단일 값 추출가능 (검색 알고리즘에 최적화됨)
# 구문은 유사함

nba.at["Austin Rivers", "Birthday"]
```

```python
nba.iat[263,1]
```

```python
# 매직 메서드
# => %% 접두사를 사용하여 선언 가능, 일반 파이썬 코드와 함께 입력함

# %%timeit : 셀에서 코드를 실행하고 실행하는데 걸리는 평균 시간을 계산하는 매직 메서드
#%%timeit 
#nba.iloc[263, 1]  # UsageError 발생
```

```python
%%timeit 
nba.iloc[263, 1]
# %%timeit 위에 다른 코드가 있으면 작동안됨. 코드블럭 맨위에 해야함.
```
